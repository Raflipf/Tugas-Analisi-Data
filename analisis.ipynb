Modul 3: Data Cleaning dan Transformasi

```python
import pandas as pd

# Membaca dataset
data = pd.read_csv("students.csv")

# Menampilkan 10 baris pertama
display(data.head(10))
```

```python
# Menampilkan informasi dataset
data.info()
```

```python
# Mengecek jumlah missing values
data.isna().sum()
```

```python
# Mengisi missing value dengan modus
data['lunch'].fillna(data['lunch'].mode()[0], inplace=True)
```

```python
# Mengisi missing value dengan mean
data['reading score'].fillna(data['reading score'].mean(), inplace=True)
```

```python
# Mengisi missing value dengan median
data['grade'].fillna(data['grade'].median(), inplace=True)
```

```python
# Mengecek informasi dataset setelah cleaning
data.info()
```

```python
# Menghapus baris dengan missing value
data.dropna(axis=0, inplace=True)
```

```python
# Menghapus kolom yang memiliki lebih dari 50% missing value
data.dropna(axis=1, thresh=int(0.5 * len(data)), inplace=True)
```
